```python
import pandas as pd

df = pd.read_csv('spotify.csv')
df
```

```python
df.info()
```

```python
# 2017 data
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce')
df['year'] = df['track_album_release_date'].dt.year

df_2017 = df[df['year'] == 2017]

df_2017
```

```python
# calculate avg
df_2017_new = df_2017.copy()
df_2017_new['duration_min'] = df_2017['duration_ms'] / 60000

average_values = df_2017_new.groupby('playlist_genre').agg({
    'loudness': 'mean',
    'mode': 'mean',
    'speechiness': 'mean',
    'acousticness': 'mean',
    'instrumentalness': 'mean',
    'liveness': 'mean',
    'valence': 'mean',
    'tempo': 'mean',
    'duration_min': 'mean',
}).reset_index()

average_values = average_values.rename(columns={'duration_ms': 'duration_minutes'})

print(average_values)

average_values.to_csv('spotify_2017.csv', index=False)

print("The average values for each genre in 2017 have been saved to 'spotify_2017_genre_averages.csv'.")
```

```python
correlation2 = df_2017_new[['danceability', 'energy']].corr()

print(correlation2)

# There is a correlation of -0.010399 between danceability and energy.
# The value is closer to 0, which indicates a weaker linear relationship between danceability and energy.
# Energy changes do not affect danceability and vice versa.
```
